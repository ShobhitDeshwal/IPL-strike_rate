```python
import pandas as pd
df = pd.read_csv('deliveries.csv')
df.head()
```

```python
df.groupby('batsman').count()
```

```python
df['ball'] = 1
df['inning'] = 1
bat_avg = df.loc[:,('batsman','inning','ball','batsman_runs')]
bat_avg = bat_avg.groupby('batsman').sum()
bat_avg['strike_rate'] =  (bat_avg['batsman_runs']/bat_avg['ball'])*100
high_sr_all = bat_avg[bat_avg['ball']>=500]
high_sr_all.sort_values('strike_rate',ascending=False)

high_sr_500 = bat_avg[(bat_avg['ball']>500) & (bat_avg['ball']< 2000) ]
high_sr_500.sort_values('strike_rate',ascending=False)

high_sr_2000 = bat_avg[bat_avg['ball']>=2000]
high_sr_2000.sort_values('strike_rate',ascending=False)

high_sr_sab = bat_avg
high_sr_sab.sort_values('strike_rate',ascending=False)
```

```python
high_sr_2000['strike_rate'] = high_sr_2000['strike_rate'].round()
high_sr_500['strike_rate'] = high_sr_500['strike_rate'].round()
high_sr_500 = high_sr_500.rename(columns={'inning':'inning_500', 'ball':'ball_500',  'batsman_runs':'batsman_runs_500'})

```

```python
import matplotlib.pyplot as plt
plt.subplot(2,2,1)
plt.hist(high_sr_2000['strike_rate'])
plt.subplot(2,2,2)
plt.hist(high_sr_500['strike_rate'])
plt.subplot(2,2,3)
plt.hist(high_sr_all['strike_rate'])
plt.subplot(2,2,4)
plt.hist(high_sr_sab['strike_rate'])
plt.xlabel('no of Players')
plt.ylabel('strike rate')
plt.show()
```

plot 1: the batman strike rate is towards the higher side though they have to start the innings and have the pressure to not loose the wickets early.
plot 2: the strike rate of strikers is at higher mean as they are supposed to strike every ball very beginning of their innings.
plot 3 : show the combined plot of strikers and the openers, the plot is almost the same which shows that the no. of strikers in a 20-20 format match are very high.
plot 4: shows the overall strike rate of all the players due to which the strike rate fall much below as it introduces the bowler and their strike rate is usually very low.
